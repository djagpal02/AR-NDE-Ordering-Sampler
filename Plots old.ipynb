```python
import numpy as np
import matplotlib.pyplot as plt
```

# True vs Random ordering

```python
# Dataset_ 1 LDBC

a = [0.9299948871135711, 0.9281080722808838, 0.9229538321495057]
b = [0.9172300517559051, 0.9174776196479797, 0.9172444343566895]

x = np.array([0,1])
y = np.array([np.mean(a),np.mean(b)])

min_max_arr = np.array([[np.min(a),np.min(b)],[np.max(a),np.max(b)]])
err = abs(min_max_arr-y)

my_xticks = ['True','Random']
plt.title('Dataset 1 - LDBC')
plt.xlabel('Ordering')
plt.ylabel('NLL')
plt.xticks(x, my_xticks)
plt.scatter(x, y)
plt.errorbar(x, y, yerr=err, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0);
plt.savefig('./Graphs/LDBC_Random.png')
plt.show()
```
